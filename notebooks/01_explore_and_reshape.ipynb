```python
import pandas as pd

path = "../data/raw/DailyNews/Combined_News_DJIA.csv"
df = pd.read_csv(path)
df.head()

df = df.drop(columns=["Label"])

headline_cols = [c for c in df.columns if c.startswith("Top")]

long_df = df.melt(
    id_vars=["Date"],
    value_vars=headline_cols,
    var_name="headline_id",
    value_name="headline_text"
)

long_df = long_df.dropna(subset=["headline_text"])
long_df.head()

long_df["ticker"] = "DJIA"

long_df.to_parquet("../data/raw/headlines_long.parquet")
```
