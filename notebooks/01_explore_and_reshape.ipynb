```python
import pandas as pd

# Load raw DJIA news
path = "data/raw/DailyNews/Combined_News_DJIA.csv"
df = pd.read_csv(path)

# Drop label (we will recompute later)
df = df.drop(columns=["Label"])

# Identify headline columns
headline_cols = [c for c in df.columns if c.startswith("Top")]

# Reshape to long format
long_df = df.melt(
    id_vars=["Date"],
    value_vars=headline_cols,
    var_name="headline_id",
    value_name="headline_text"
)

# Drop empty headlines
long_df = long_df.dropna(subset=["headline_text"])

# Add ticker
long_df["ticker"] = "DJIA"

# Save as parquet
long_df.to_parquet("data/raw/headlines_long.parquet", index=False)

# Reload to verify
df_check = pd.read_parquet("data/raw/headlines_long.parquet")
df_check.head(), df_check.shape
```
